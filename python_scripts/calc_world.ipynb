```python
import pandas as pd
import zipfile
import os
```

```python
files = ['stats', 'stops.geojson']
prefix = '../data/world_data/'
for file in os.listdir('../data/world_data/1186215/'):
    os.mkdir(f'{prefix}/{file.split(".")[0]}')
```

```python
output_df = pd.DataFrame(columns=['city','center_lat','center_long','network_length'])
path = '../data/world_data/'
for folder in os.listdir(path):
    df = pd.read_csv(path+folder+'/stats.csv', delimiter=';')
    center_lat = df['buffer_center_lat'].item()
    center_long = df['buffer_center_lon'].item()
    network_len = df['network_length_m'].item()
    output_df.loc[len(output_df.index)] = [folder.capitalize(),center_lat,center_long,network_len]
```

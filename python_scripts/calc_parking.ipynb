```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../data/On-street_Car_Parking_Sensor_Data_-_2019.csv')
```

```python
df.head()
```

```python
v_pres = df[df['VehiclePresent'] == True]
```

```python
v_pres.head()
v_pres = v_pres[['ArrivalTime', 'DepartureTime', 'DurationMinutes', 'InViolation', 'VehiclePresent']]
```

```python
v_pres['Time'] = v_pres['ArrivalTime'].apply(lambda x: ' '.join(x.split(' ')[1:]))
v_pres.head()
```

```python
v_pres['arrival_parsed'] = pd.to_datetime(v_pres['ArrivalTime'], format="%m/%d/%Y %I:%M:%S %p")
```

```python
v_pres.head()
```

```python
v_pres['day_of_week'] = v_pres['arrival_parsed'].dt.day_of_week
v_pres['day_of_week'] = v_pres['day_of_week'].map({
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
})
v_pres.head()
```

```python
import numpy as np
bins = np.arange(25)
v_pres['Hour_Pres'] = pd.cut(v_pres['arrival_parsed'].dt.hour, bins, right=False)
v_pres['Hour_Pres'] = v_pres['Hour_Pres'].map(lambda x: x.left)
```

```python
v_pres.head()
```

```python
counts = v_pres[['day_of_week','Hour_Pres']].value_counts(sort=False)
counts
```

```python
output_df = v_pres.groupby(by=["day_of_week", "Hour_Pres"]).size().reset_index(name="Count")
output_df['Average'] = output_df['Count'].apply(lambda x: x/365)
```

```python

output_df.to_csv('../data/parking_times.csv', index=False)
```
